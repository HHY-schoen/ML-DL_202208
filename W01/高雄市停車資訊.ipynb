### 高雄市停車資訊

- 不是每一次資料都是沒問題的....

```python
import requests

url = "https://data.kcg.gov.tw/dataset/449e45d9-dead-4873-95a9-cc34dabbb3af/resource/fe3f93da-9673-4f7b-859c-9017d793f798/download/108.6.21.csv"
r = requests.get(url)
print(r)
```

```python
import csv

decoded_content = r.content.decode('utf-8-sig')
cr = csv.reader(decoded_content.splitlines(), delimiter=',')
data_list = list(cr)
```

```python
print(data_list[0])
print(data_list[1])
```

```python
#更好的寫法
with open('kh_parking.csv', 'r', encoding="utf-8-sig") as f:
    cr = csv.reader(f)
    data_list = list(cr)
data_list[0:2]
```

```python
import folium
import geocoder

# 使用 geocoder 取得特定住址的 GPS 座標
location = geocoder.osm('高雄市').latlng
print(location)
```

```python
print(data_list[0])
print(data_list[1])
```

```python
m = folium.Map(location=location, zoom_start=15)

for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        name = item[2]  #站名
        lat = item[5]  #緯度
        lng = item[4]  #經度
        car_n = item[7]
        fee = item[9]
        info = '[%s] 小型車格數=%s, 收費=%s' %(name, car_n, fee)
        folium.Marker([ float(lat), float(lng)],
                      tooltip=info,
                     icon=folium.Icon(color='blue', prefix='fa', icon='fa-car')
                     ).add_to(m)
        
    except Exception as e:
        print(e.args)
```

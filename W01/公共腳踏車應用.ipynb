### 新北市公共自行車租賃系統(YouBike): https://data.gov.tw/dataset/123026

```python
import requests
```

```python
url = "https://data.ntpc.gov.tw/api/datasets/71CD1490-A2DF-4198-BEF1-318479775E8A/csv/file"
r = requests.get(url)
r
```

```python
decoded_content = r.content.decode('utf-8-sig')
print(decoded_content)
```

```python
# 用csv套件解析
import csv

cr = csv.reader(decoded_content.splitlines(), delimiter=',')
data_list = list(cr)
```

```python
print(data_list[0])
```

```python
print(data_list[1])
```

```python
import folium
import geocoder

# 使用 geocoder 取得特定的 GPS 座標
location = geocoder.osm('新北市').latlng
print(location)
```

```python
m = folium.Map(location=location, zoom_start=15)

for item in data_list[1:]:
    try:
        # item[2]=緯度, item[1]=經度, item[0]=名稱
        name = item[1]  #站名
        total = item[2] #可借總數量
        sbi = item[3]  #剩餘數量
        lat = item[6]  #緯度
        lng = item[7]  #經度
        info = '%s 剩餘數量[%s/%s]' %(name, sbi, total)
        folium.Marker([ float(lat), float(lng)],
                      tooltip=info,
                     icon=folium.Icon(color='blue', prefix='fa', icon='fa-bicycle')
                     ).add_to(m)
        
    except Exception as e:
        print(e.args)    
```

```python
m
```

- 進階應用

```python
m = folium.Map(location=location, zoom_start=14)

for item in data_list[1:]:
    try:
        name = item[1]
        area = item[4]
        total = item[2]
        n = item[3]
        lat = item[6]
        lng = item[7]
        if int(n)<5 and int(n)>0:
            folium.Marker([float(lat), float(lng)], popup=name+'<br>剩餘車輛:%s/%s' %(n,total),
                        icon=folium.Icon(color='red', prefix='fa', icon='fa-bicycle')).add_to(m)
        elif int(n)==0:
            folium.Marker([float(lat), float(lng)], popup=name+'<br>剩餘車輛:%s/%s' %(n,total),
                        icon=folium.Icon(color='black', prefix='fa', icon='fa-bicycle')).add_to(m)
        else:
            folium.Marker([float(lat), float(lng)], popup=name+'<br>剩餘車輛:%s/%s' %(n,total),
                        icon=folium.Icon(color='green', prefix='fa', icon='fa-bicycle')).add_to(m)
        
    except Exception as e:
        print(e.args)  
```

```python
m
```
